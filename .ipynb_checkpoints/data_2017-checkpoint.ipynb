```python
credits = pd.read_csv('datasets/credits.csv')
```

```python
#parsing dates properly
meta['release_date'] =  pd.to_datetime(meta['release_date'], errors='coerce' )
```

```python
#meta.release_date.dt.year.value_counts().sort_index()
meta['year'] = meta['release_date'].dt.year
```

```python
new_meta = meta.loc[meta.year == 2017, ['genres', 'id', 'title', 'year']]
```

```python
new_meta
```

```python
new_meta.dtypes
```

```python
new_meta['id'] = new_meta.id.astype('int')
```

```python
new_meta.dtypes
```

```python
#merging the files
new_data = pd.merge(new_meta, credits, on='id')
```

```python
new_data
```

```python
#using ast library
import ast
```

```python
new_data['genres'] = new_data['genres'].map(lambda x: ast.literal_eval(x))
new_data['cast'] = new_data['cast'].map(lambda x: ast.literal_eval(x))
new_data['crew'] = new_data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
new_data.genres[0][0].values()
```

```python
def make_genreList(x):
    gen = []
    st  = ' '
    for genre in x:
        if genre.get('name') == 'Science Fiction':
            gen.append('Sci-Fi')
        else:
            gen.append(genre.get('name'))
        
    if gen == []:
        return np.NaN
    else:
        return st.join(gen)
```

```python
new_data['genres_list'] = new_data['genres'].map(lambda x: make_genreList(x)) 
```

```python
new_data
```

```python
def get_actor_names(cast):
    casts = []
    for actor in cast:
        casts.append(actor.get('name'))
        if len(casts) == 3:
            break
    return casts
```

```python
new_data['actor_1_name'] = np.NaN
new_data['actor_2_name'] = np.NaN
new_data['actor_3_name'] = np.NaN
```

```python
new_data
```

```python
#adding actor names to df
pd.set_option('mode.chained_assignment', None)
for i in new_data.index:
    cast_list = get_actor_names(new_data['cast'][i])
    if len(cast_list) == 3:
        new_data['actor_1_name'][i] = cast_list[0]
        new_data['actor_2_name'][i] = cast_list[1]
        new_data['actor_3_name'][i] = cast_list[2]
    elif len(cast_list) == 2:
        new_data['actor_1_name'][i] = cast_list[0]
        new_data['actor_2_name'][i] = cast_list[1]
    elif len(cast_list) == 1:
        new_data['actor_1_name'][i] = cast_list[0]
```

```python
new_data
```

```python
def get_directors(crew):
    directors = []
    for director in crew:
        if director.get('job') == 'Director':
            directors.append(director.get('name'))
    if directors == []:
        return np.NaN
    else:
        return  (' ').join(directors)      
    
```

```python
new_data['director_name'] = new_data['crew'].map(lambda x: get_directors(x))
```

```python
new_data
```

```python
final_data = new_data[['director_name', 'actor_1_name', 'actor_2_name', 
                       'actor_3_name', 'genres_list', 'title']]
```

```python
final_data
```

```python
final_data.isna().sum()
```

```python
final_data = final_data.dropna(how='any')
```

```python
final_data
```

```python
final_data.rename(columns={'genres_list': 'genres', 'title': 'movie_title'}, inplace=True)
```

```python
final_data
```

```python
final_data['movie_title'] = final_data['movie_title'].str.lower()
```

```python
final_data['summary'] = final_data['actor_1_name'] + ' ' + final_data['actor_2_name'] + ' '+ \
final_data['actor_3_name'] + ' '+ final_data['director_name'] +' ' + final_data['genres']
```
